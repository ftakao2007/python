# 社会保険

* 大きく分けて5種類
    * 健康保険
        * ケガや病気で病院に行った際に、病院に提出する健康保険証の交付
        * 中小企業であれば「全国健康保険協会」に加入するのが一般的
    * 厚生年金保険
        * 企業に勤めている人や公務員が加入する年金制度
    * 労災保険
        * 業務中や通勤中の事故や災害により、従業員がケガをしたり、病気になったり、体に障害が残ったり、死亡したりした場合に備える制度
    * 雇用保険
        * 失業してしまったときの生活や雇用の安定、就職の促進のための制度
    * 介護保険
        * 40歳以上の方は、上の4種類に加えて介護保険にも加入しなければならない
        * 自分の介護が必要になったときに、介護にかかる費用を最大9割負担してもらえる

## 扶養

* 配偶者
    * 妻、もしくは夫
        * 配偶者控除
            * 配偶者の給与収入が103万円以下の場合に、年末調整で本人の税金（所得税・住民税）を一定額減らすことができる制度
        * 配偶者特別控除
            * 配偶者に103万円を超える収入があるため配偶者控除の適用が受けられないときに適用される
            * 配偶者の所得金額に応じて、一定の金額の所得控除が受けられる
                * 年収が103万円超え～130万円までの間

* 扶養親族
    * 配偶者以外の親族（6親等内の血族及び3親等内の姻族）
        * ただし、控除を受けられるのは16歳以上
    * 納税者と生計を一にしている
    * 年間の合計所得金額が38万円以下であること(給与のみの場合は給与収入が103万円以下)
    * 青色申告者の事業専従者としてその年を通じて一度も給与の支払を受けていないこと又は白色申告者の事業専従者でないこと

* 扶養の種類
    * 税法上の扶養
        * 扶養者(配偶者、子ども)の給与年収が103万円以下である場合に入ることが可能
        * 被扶養者(ex.妻)は、本来支払わなければならない所得税や住民税の一部が免除される
        * 103万円を超えてしまうと扶養に入ることができないので、個人で所得税や住民税の税金を納めなければならない
        * 2018年1月より、配偶者控除の上限が103万円から150万円に変更になった
            * 配偶者以外の扶養者はこれまで通り103万円なので注意
    * 社会保険上の扶養
        * 健康保険の扶養となる人は、自分で保険料を支払うこと無く、健康保険に加入することができる
        * 厚生年金保険の扶養となる人は、国民年金保健に加入することができる
        * 条件
            * 年収が130万円未満
            * その家族は健康保険法に定める被扶養者の範囲である
                * 父、母、子供の場合は同居してなくても良い
            * 後期高齢者に該当していない
                * 75歳以上
            * 被保険者がその家族を扶養せざるを得ない理由がある
            * 被保険者がその家族を経済的に主として扶養している事実がある
                * その家族の生活費を主として負担している
            * 被保険者には継続的にその家族を養う経済的扶養能力がある
            * その家族の年収は被保険者の年収の1/2未満
            * その家族の収入は年間130万円未満（60歳以上又は59歳以下の障害年金受給者は年間180万円未満）

## 健康保険・厚生年金保険

* 年金事務所へ提出
    * 健康保険・厚生年金保険新規適用届
        * 登記簿謄本原本
    * 健康保険・厚生年金保険被保険者資格取得届
    * 健康保険被扶養者（異動）届
        * 被扶養者の健康保険被保険者
        * 扶養者の年間所得が103万円以上130万円未満の場合は、「課税（非課税）証明書」の添付も必要
* 保険料
    * 「標準報酬月額」(加入者の毎月のお給料)で保険料が決まる
   

## 労働保険

* 「雇用保険」と「労災保険」を総称して労働保険と呼ぶ

## 雇用保険

* 従業員を雇ったときに加入
    * 週の所定労働時間が20時間以上、且つ31日以上雇用の継続が見込まれる従業員（パートやアルバイトも含む）を1人でも雇えば加入
* 保険料は事業主と従業員が負担
* ハローワークに提出
    * 雇用保険適用事業所設置届
    * 雇用保険被保険者資格取得届

## 労災保険

* 従業員を雇った場合に加入
    * 従業員（パートやアルバイトも含む）を1人でも雇えば加入
* 保険料は全額事業主が負担
* 労働基準監督署に提出
    * 労働保険関係成立届
        * 登記謄本原本、労務者名簿、賃金台帳、出勤簿
    * 労働保険概算保険料申告書
