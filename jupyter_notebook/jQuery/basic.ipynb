# jQuery

* jQuery
    * ウェブブラウザ用のJavaScriptコードをより容易に記述できるようにするために設計された軽量なJavaScriptライブラリ
* JavaScript
    * HTMLとCSSを操作するプログラミング言語
    * 以下を強く意識することが重要
        * 操作する要素(の属性)
        * 操作する内容
        * 操作するタイミング
    * HTMLの例
        * `<a href="http://hijiriworld.com/web/" target="_blank">リンク</a>`
            * 要素 : a
            * 属性 : href, target
            * 値
                * herfの値 `http://hijiriworld.com/web/`
                * targetの値 `_blank`
    * CSSの例
        * `a { color: red; }`
            * 要素 : a
            * プロパティ(属性) : color
            * 値 : red

## イベント

* イベント
    * 操作するタイミング
* readyイベント
    * ページ構成（DOM要素）を読み込んだ時点で処理する
        * 画像などページ全体を読み込みではない
            * JavaScriptのonloadイベントは、例えば画像を読み込み終わってから処理する
        * ページ全体を読み込む前に様々な準備ができる
    * ほとんどのjQueryコンテンツで必要
    * readyイベントを利用しないと
        * <font color="red">DOM要素が読み込まれる前にメソッドを実行してしまう</font>
            * HTML内の要素を操作する場合はreadyイベントの仕組みを利用しないと正しい結果が得られない
* 構文
    * 簡単な構文
        * `$(セレクタ).イベント名(イベントが発生したときに実行したい処理)`
            * イベントが発生したときに実行したい処理はfunctionで設定
        * <font color="red">動的な要素には対応できない</font>
    * onメソッドを利用した構文
        * `調査範囲.on(イベント名, セレクタ, イベントで実行したい処理)`
            * <font color="red">調査範囲はjQueryオブジェクト</font>にする
            * イベント名とセレクタは<font color="red">クオートで囲む</font>
                * イベント名は文字として指定するため<font color="red">頭のドットと括弧は不要</font>
                    * .click() -> "click"
                * セレクタは文字列で指定するため<font color="red">jQueryオブジェクトへの変換は不要</font>
            * イベントで実行したい処理はfunctionで設定

## セレクタ
   
* セレクタ
    * CSSにおいて対象を指定する仕組み
        * <font color="red">jQueryの仕組みではない</font>
    * セレクタからjQueryオブジェクトを生成する
        * img要素をjQueryで扱えるようにjQueryオブジェクトを生成する例
            * `$("img");`
                * セレクタの場合は<font color="red">クオートで囲む</font>

## オブジェクト

* オブジェクト
    * JavaScriptのオブジェクト
        * <font color="red">「プロパティ」と「メソッド」を持つ構造体</font>
        * スクリプトで<font color="red">操作する対象</font>のことを指す
            * プロパティ
                * 特定のデータが格納されている変数みたいなもの
                * 例) navigator.userAgent
                    * navigatorオブジェクトのuserAgentプロパティに格納されている値を取得
                * 書式
                    * オブジェクト.プロパティ
            * メソッド
                * オブジェクトがもっている関数
                * 例) history.back()
                    * historyオブジェクトのbackメソッドを実行
                * 書式
                    * オブジェクト.メソッド()
    * jQueryオブジェクト
        * jQueryの命令をを持ったオブジェクト
        * 生成方法
            * ` jQuery(セレクタ) `
            * ` $(セレクタ) `
    * DOM要素
        * 他の呼び方
            * DOMオブジェクト
        * DOM要素からjQueryオブジェクトを生成する
            * documentからjQueryオブジェクトを生成する例
                * `$(document).ready( //処理 );`
                    * DOM要素の場合は<font color="red">クオートで囲まない</font>
                    * 略される事がほとんどなので利用頻度はそれほど高くない
                    * windowというブラウザのウインドウを指すDOM要素もある
        * this
            * JavaScriptの仕組み
                * <font color="red">jQueryの仕組みではない</font>
            * .<font color="red">イベントで設定したfunction内で利用</font>すると、イベントが発生した要素を指す
                * このときthisは `$( )` 内で<font color="red">クォートで囲まない</font>
                * thisを指定しないと、例えばあるdiv要素にある操作を設定したとき、その要素だけではなく他の要素も操作されてしまう
                * <font color="red">イベント以外にもthisは使われる</font>のでそのthisとは別で考える

## メソッド

* メソッド
    * オブジェクトを操作する命令文
    * 構文
        * `オブジェクト.メソッド`

## 情報の取得

```
変数 = $(セレクタ).メソッド(引数)
変数 = オブジェクト.プロパティ
```

## 動くものを見る

* 以下のコードをhtmlファイルとして保存してブラウザで見ると「おばけー」という文字が表示され、クリックすると消える

```
<!DOCTYPE HTML PUBLIC "-//W3C//DTD HTML 4.01 Transitional//EN">
<meta http-equiv="content-type" content="text/html; charset=UTF-8">
<html>
  <head>
    <meta charset="utf-8">
    <script src="http://ajax.googleapis.com/ajax/libs/jquery/1.7.0/jquery.min.js"></script>
    <title>おばけ</title>
    <script>
      $(function() {
        $('p').click(function() {$(this).fadeOut('slow');});
      });
    </script>
  </head>

  <body>
    <p>おばけー</p>
  </body>
</html>
```

````python
%%bash
### 必要な設定の読み込み
source ${HOME}/secrets
name=jquery.html

### ソースの編集
cat <<- 'EOS' > ${name}
    
<!DOCTYPE HTML PUBLIC "-//W3C//DTD HTML 4.01 Transitional//EN">
<html>
  <head>
    <meta http-equiv="content-type" content="text/html; charset=UTF-8">
    <meta charset="utf-8">
    <script src="http://ajax.googleapis.com/ajax/libs/jquery/1.7.0/jquery.min.js"></script>
    <title>おばけ</title>
    <script>
      $(function() {
        $('p').click(function() {$(this).fadeOut('slow');});
      });
    </script>
  </head>

  <body>
    <p>おばけー</p>
  </body>
</html>

EOS

### ソースの転送
sshpass -p "${pass}" scp ${name} ${user}@${host}:${path}
echo "created ${host}:${path}/${name}"
````

## 解説

```
================================================
<script>
  $(function() {
    $('p').click(function() {$(this).fadeOut('slow');});
  });
</script>
================================================
```

省略せずに書くと以下のようになる

```
================================================
<script>
// DOMの準備が終わった時点でsetReady関数を呼び出す
jQuery(document).ready(setReady);

// p要素がクリックされたときにsetClick関数を呼び出す
function setReady() {
  jQuery('p').click(setClick);
}

// 押された要素をゆっくりとフェードアウトする
// thisはイベントが発生した要素(クリックされたp要素)
function setClick() {
  jQuery(this).fadeOut('slow');
}
</script>
================================================
```

* DOM(Document Object Model)
    * HTMLを扱うための仕組み
    * 特徴
        * ツリー構造とも呼ばれる階層構造を取る
            * 例 : `<body><section><p></p></section><section><p></p></section></body> `
            * この階層構造を定義しているものこそがDOMと呼ばれている仕組み
        * それぞれノードという言葉で説明される
            * 先ほどの例で1つめの`<section>`を基準ノードとしたとき
                * `<body>` : 親ノード
                * 1つめの `<p>` : 子ノード
                * 2つめの `<section>` : 兄弟姉妹ノード
            * jQueryでは「そのタグの子ノード全体を取得して、その親ノードから削除する」のような使い方ができる
        * WEBページとJavaScriptなどのプログラミング言語とを繋ぐ
            * DOMで定義されている「ノード」などの仕様を通じてJavaScriptからHTMLを操作することができる
* jQueryは多くの処理を `jQuery(document).ready` 内に記述する
    * そうしないと、body内の要素を使える保証が無くなる



## jQueryの別名による省略方法

* 「jQuery」という識別子の別名
    * 「`$`」
    
```
jQuery(document).ready(...)
↓
jQuery(...)
↓
$(...)
```

## 無名関数による省略法

* 関数を引数に取る関数があったとき、引数内に関数を直接書き、関数名を省略することができる

```
func_a(func_b);
function func_b() {処理}
↓
func_a(function() {処理});
```


## jQueryの基本構文

```
================================================
$(セレクタ).イベント(function() {
    $(セレクタ).命令
});
================================================
```

* セレクタ
    * 操作する要素を指定する
* 命令
    * 操作する内容
* イベント
    * 操作するタイミング
    
## プラグインの基本構文

* メソッド名を決めて「 jquery.メソッド名.js」というファイルを作成する
* プラグイン内のセレクターは、イベントが発生した要素を取得する `$(this)` にしておく

```
###  jquery.メソッド名.js
================================================
$.fn.メソッド名 = function() {
    //ここに処理を書く
}
================================================
```

* 上のファイルを読み込んでメソッドを呼び出せばOK
* オプションとして引数も設定できる

```
================================================
$(function(){
    $(セレクター).メソッド名();
});
================================================
```

# タグ名、class、ID

* タグ
    * どのようなページなのかという情報や、ページ上のテキスト、画像、リンクなどを表示する際に必要なもの
    * 例
        * `<html>, <body>, <font>` など
* ID
    * 定義
        * この属性は、要素に名前を割り当てる。この名前は<font color="red">文書中で一意でなければならない。</font>
    * 名札みたいなもの
        * 同じ場所に同じ名札の人が居ると混乱する
* class
    * 定義
        * この属性は、ある要素に1つのクラス名を割り当てるか、または複数のクラス名を設定する。
        * <font color="red">幾つの要素に対してでも、同じクラス名あるいはクラス名群が割り当てできる。</font>
        * 複数のクラス名については、空白文字によって区切らねばならない。
    * グループ名みたいなもの
