## data.toString()の概要

```python
%%bash
source ~/.nvm/nvm.sh
html=jquery_tmp.html
js=tmp.js

### jsファイルの作成
### ヘッダ部分
cat <<- 'EOS' > ${js}
const jsdom = require("jsdom");
const { JSDOM } = jsdom;
const { window } = new JSDOM(`<!DOCTYPE html>`);
const $ = require('jquery')(window);
const fs = require('fs');

EOS

### htmlファイル指定部分
cat <<- EOS >> ${js}
fs.readFile(__dirname + "/${html}", function (err, data) {
  if (err) throw err;

EOS

### スクリプト部分
cat <<- 'EOS' >> ${js}

  $(function() {
    var body = data.toString();
    //console.log(body);
    //console.log("================================================");
    console.log($(body));
    console.log("================================================");
    console.log($(body).text());
    console.log("================================================");
    console.log($(body)[0]);
    console.log($(body)[1]);
    console.log($(body)[2]);
    console.log($(body)[3]);
    console.log($(body)[4]);
    console.log($(body)[5]);
    console.log($(body)[6]);
    console.log($($(body)[7]).text());
    console.log($(body)[8]);
    console.log($(body)[9]);
    console.log($(body)[10]);
    console.log($(body)[11]);
    console.log($(body)[12]);
    console.log($(body)[13]);
    console.log($(body)[14]);
});
EOS

### fs.readFile閉じ
cat <<- 'EOS' >> ${js}

});
EOS

### 実行
node ${js}
```

## それぞれの要素の取得

```python
%%bash
source ~/.nvm/nvm.sh
html=jquery_tmp.html
js=tmp.js

### jsファイルの作成
### ヘッダ部分
cat <<- 'EOS' > ${js}
const jsdom = require("jsdom");
const { JSDOM } = jsdom;
const { window } = new JSDOM(`<!DOCTYPE html>`);
const $ = require('jquery')(window);
const fs = require('fs');

EOS

### htmlファイル指定部分
cat <<- EOS >> ${js}
fs.readFile(__dirname + "/${html}", function (err, data) {
  if (err) throw err;

EOS

### スクリプト部分
cat <<- 'EOS' >> ${js}

  $(function() {
    var body = data.toString();
    
    var result_h2 = $(body).find('h2')
    var result_h2_text = $(body).find('h2').text()
    var result1 = $(body).find('h2')['length']
    var result2 = $(body).find('h2')['prevObject'].text()
    
    console.log(result_h2);
    console.log("================================================");
    console.log(result_h2_text);
    console.log("================================================");
    console.log(result1);
    console.log("================================================");
    console.log(result2);
  });
EOS

### fs.readFile閉じ
cat <<- 'EOS' >> ${js}

});
EOS

### 実行
node ${js}
```
