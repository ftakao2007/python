## each()メソッド

* 複数の要素に対して処理を行う

### 基本構文

```
================================================
$( セレクタ ).each(function(インデックス, エレメント) { 
  ループ処理
});
================================================
```


### section-1のp要素の中のテキストを取得(引数無し)

* `$(this)` を使って要素を取得

```
================================================
$('#section-1 p').each(function() {
    console.log($(this).text());
});
===
子ノード1
子ノード2
子ノード3
================================================
```

### 引数を指定して処理

* 第一引数 i : index
* 第二引数 `$(val)` : 値

```
================================================
$('#section-1 p').each(function(i, val) {
    console.log(i+':'+$(val).text());
});
===
0:子ノード1
1:子ノード2
2:子ノード3
================================================
```

### 第一引数のみ指定した処理

* 値は `$(this)` で取得

```
================================================
$('#section-1 p').each(function(i) {
    console.log(i+':'+$(this).text());
});
===
0:子ノード1
1:子ノード2
2:子ノード3
================================================
```


## JQuery.each()メソッド

* `$.each()` メソッドともかける
* 普通の配列やjQueryオブジェクトでないものも扱える
    * 普通のeach()メソッドはjQueryオブジェクトに対してのみ使用できる

### 基本構文

```
================================================
$.each( 配列やオブジェクト , function(インデックス（オブジェクトはキー）, エレメント) { 
  ループ処理
});
================================================
```

### 要素をループ処理

* 先ほどのeach()と同じ処理

```
================================================
$.each($('#section-1 p'), function(i, val) {
    console.log(i+':'+$(val).text());
});
===
0:子ノード1
1:子ノード2
2:子ノード3
================================================
```

### 配列を処理

* each()と違い配列を処理できる

```
================================================
var arr = [ "apple", "orange", "grape" ];
$.each(arr, function(i, val) {
    console.log(i+':'+val);
});
===
0:apple
1:orange
2:grape
================================================
```
