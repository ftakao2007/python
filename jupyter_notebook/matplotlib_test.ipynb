```python
import matplotlib.pyplot as plt
import numpy  as np
%matplotlib inline
```

```python
### figureインスタンスの作成
# デフォルト
fig = plt.figure()

# 画面大きさを指定
#fig = plt.figure(figsize=(5,5))
#fig = plt.figure(figsize=(10,10))

### 軸を追加
ax = fig.add_subplot(1,1,1)
#ax = fig.add_subplot(2,2,2)
```

# y = x のグラフ

```python
### xが0から10の間で0.1刻み
x = np.arange(0, 10, 0.1)

y = x
plt.plot(x, y)
```

# タイトル付き

```python
x = np.arange(0, 5, 0.1)
y = np.exp(x)

### タイトルをつける
plt.title("exponential function: $ y = e^x $")

plt.plot(x, y)
```

# Y軸の表示範囲設定

```python
x = np.arange(0, 5, 0.1)
y = 10*x**2
plt.title("exponential function: $ y = x^2 $")

### Y軸の表示範囲を0〜100に設定
plt.ylim(0, 100)

plt.plot(x, y)
```

# 破線を入れる

```python
### xの範囲の最大と最小
xmin, xmax = -np.pi, np.pi
x = np.arange(xmin, xmax, 0.1)

y_sin = np.sin(x)
y_cos = np.cos(x)
plt.plot(x, y_sin)
plt.plot(x, y_cos)

###  y=-1, 1に破線を描画
plt.hlines([-1, 1], xmin, xmax, linestyles="dashed")

plt.title("$\sin(x)$ and $\cos(x)$")
plt.xlim(xmin, xmax)
plt.ylim(-1.3, 1.3)
```
