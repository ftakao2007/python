## ファイナンシャル・プランニング

* ライフデザイン
    * 将来の人生計画
        * 結婚、子育て、住居、老後の暮らしなどについてその人の価値観に応じて思い描くこと
* ライフプラン(人生設計)
    * ライフデザインで思い描いた内容を具体化した生涯の人生設計
        * ファイナンシャル・プランニング
            * 経済的な面を中心にライフプランを立てること
        * ファイナンシャル・プランナー(FP)の役割
            * ファイナンシャル・プランニングの設計や実行、見直しの支援を行う
* ファイナンシャル・プランニングの手順
    1. 顧客との関係確立(信頼関係を作る)
    1. 顧客の情報収集と目標の明確化(目的や希望の確認)
    1. 現状分析(資金面などの問題点の洗い出し)
    1. 提案書の作成と定時
    1. プランの実行と支援
    1. プランの定期的なフォロー(見直し)
* FPの職業倫理
    * 顧客利益の優先
        * FP自身や第三者の利益を優先してはいけない
    * 顧客情報の守秘義務
        * 顧客のプライバシーに関わる情報を顧客の同意を得ずに外部に漏らしてはいけない    

## FP業務の関連法規

FPは税理士法や弁護士法、保険業法、金融商品取引法などに抵触しないようにするとともにこれらの分野の専門家と連携して業務を行うことが求められる

### 税理士法

FPは(1)を問わず、個別の具体的な(2)や(3)(確定申告書など)をおこなってはならない

例外

仮の事例を用いての税金の計算や、(4)な税金の說明を行うことは可能

2. 税務相談

4. 一般的

### 弁護士法

FPは(1)および(2)(遺言書の作成など)を受けてはならない

例外

弁護士資格などを持っていないFPでも、成年後見制度における(3)にはなることができる

<!--
1. 法律相談
2. 法律事務
3. 任意後見人
-->

### 保険業法

保険の(1)を行うには(2)として内閣総理大臣の(3)を受けなければならず、FPは保険の(1)を行ってはならない

例外

将来の必要保障額の計算や、(3)、保険の見直しなどの相談に応じることは可能

<!--
1. 募集・勧誘・販売
2. 保険募集人
3. 保険商品の說明
-->

### 金融商品取引法

有価証券投資の助言や代理業等を行うには(1)(証券会社のこと)として内閣総理大臣の(2)を受けなければならない

例外

新聞や雑誌などを用いて景気動向や企業分析など、投資判断の前提となる(3)な情報を知らせることは可能

<!--
1. 金融商品取引業者
2. 登録
3. 一般的
-->
