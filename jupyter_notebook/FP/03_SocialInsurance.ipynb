# 社会保険制度

* 社会保険制度の種類
    * 医療保険制度
        * 国民皆保険制度
            * 国内に住所のあるものは、以下のいずれかの医療保険に加入する
                * (1) 健康保険
                    * 協会けんぽ(全国健康保険協会管掌健康保険)
                        * 全国健康保険協会が運営
                        * 主に中小企業の従業員とその家族
                    * 組合健保(組合管掌健康保険)
                        * 健康保険組合が運営
                        * 主に大企業の従業員とその家族
                * (2) 国民健康保険
                    * 自営業者などが対象
                * (3) 後期高齢者医療制度
                    * 原則として75歳以上の高齢者が対象
                    * 65歳以上の障害者も対象
                    
    * 介護保険制度
    * 労働者災害補償制度(労災保険)
    * 雇用保険制度
    * 年金保険制度

## 健康保険

* 目的
    * <font color="red">業務外</font> の病気や怪我、出産、死亡に対して保険金を給付する
        * 業務上の場合は原則として労災保険
* 被保険者(対象者)と被扶養者
    * 企業などで働く役員や従業員
    * 被扶養者(<font color="red">3親等内の親族など</font>)
        * 被保険者に扶養されている家族のこと。被扶養者は保険料を収める必要はない。
        * 被扶養者になれる者
            * 年収130万円未満 (60歳以上や障害者は180万円未満)
            * 被保険者の年収の1/2未満
        * 以下の条件を<font color="red">全て満たした</font>パートなどの短期労働者(学生を除く)は健康保険や年金制度の被扶養者ではなくなる
            1. 月収が8万8000円(通常年収106万円)以上
            1. 1年以上雇用見込み
            1. 週の労働時間が20時間以上
            1. 従業員数が501人以上の企業で働いている
* 保険料
    * 事業主と従業員が半分ずつ負担する(<font color="red">労使折半</font>)
    * 標準報酬月額(給料の平均)と標準賞与額(ボーナスの平均)に保険料率をかける
        * 保険料率
            * 協会けんぽ : 都道府県ごとに異なる
            * 組合健保 : 各組合の規約で定める
    * 40歳以上65歳未満は<font color="red">介護保険料</font> をあわせて支払う
    * 育児休業中や産休中の健康保険料や厚生年金保険料は被保険者、事業主共に免除される
* 主な給付内容
    * 療養の給付
        * <font color="red">業務外</font> の病気や怪我に対する給付
        * 外来でも入院でも負担率は同じ
        * 医療費の一部は自己負担になる
            * 69歳以下 : 3割負担。ただし小学校入学前は2割負担。
            * 70〜74歳 : 2割負担。現役並みの所得者は3割負担。
            * 75歳以上 : 1割負担。現役並みの所得者は3割負担。
    * 高額療養費の給付
        * <font color="red">1ヶ月間(同じ月であること)</font> に同一の診療を受け、自己負担額が一定基準を超えた場合にその超過額が支給される制度
        * 事前に「自己負担限度額適用認定申請書」を提出しておけば窓口での支払額は自己負担限度額までになる
            * 
    * 傷病手当金
    * 出産育児一時金
    * 出産手当金
    * 埋葬料
    
    

## 国民健康保険

* 被保険者(対象者)
    * 自営業者や定年退職社
    * 被扶養者という制度はない
* 保険者
    * 2種類
        * 市区町村
        * 国民健康保険組合(同業種の個人を対象とするもの)
* 健康保険との違い
    * 傷病手当金無し
    * 出産手当金無し
    * 原則として業務上のけがや病気についても保険金が給付される

## 後期高齢者医療制度

# 退職後の医療保険制度

* 会社員の場合以下3つから選ぶ
    * <font color="red">任意継続被保険者</font>となって引き続き勤めていた会社の健康保険に加入する
        * 原則として傷病手当金や出産手当金は受けられなくなる
        * 加入資格 : 被保険者期間が継続して2ヶ月以上ある者
        * 保険料 : 全額自己負担
        * 申請期間 : 退職日の翌日から20日以内
        * 加入期間 : 2年間
    * 子や配偶者などの被扶養者になる
        * 被扶養者になる要件(収入など)を満たしている
        * 保険料 :  不要
        * 申請期間 : すみやかに申請する
        * 加入期間 : 制限なし
    * 自ら国民健康保険に入る
        * 退職日の翌日から14日以内に申請する


# 介護保険制度

# 労働者災害補償保険 (労災保険)

* 概要
    * 労働者が業務上又は通勤中に、負傷、疾病、障害、死亡した場合に保険給付を行う
    * 業務上災害と通勤災害が対象になる
    * 原則自己負担なし
        * 治療費は全額労災保険より支給される
    * 被保険者
        * 労働者は自動的に全員が被保険者となる
        * ただし経営者は対象外
    * 窓口
        * 労働基準監督署
    * 保険料
        * 全額事業主負担
            * 保険料率は業種により異なる
    * 特別加入制度
        * 特別に加入できる制度
            * 一定の中小企業の事業主
            * 個人タクシー
            * 大工などの一人親方
            * 海外の事業所で働く者
* ポイント
    * 業務災害
        * 業務命令による出張中の事故は全て該当
    * 通勤災害
        * 通常の経路で自宅と会社の間を往復する途中に災害にあった場合
        * 寄り道は原則対象外
            * ただし、日用品の購入など日常生活に必要止む得ない場合は対象となる
* 休業補償給付
* 療養補償給付
* 障害補償給付
* 傷病補償年金

# 雇用保険

* 概要
    * 失業時の給付
    * 再就職の支援や失業者に対する職業訓練のための給付
    * 雇用の継続に必要な給付
    * 適用事業所
        * 原則、労働者を雇用するすべての適用事業所
    * 被保険者(対象者)
         * 原則、すべての労働者
             * パート社員や派遣労働者は一定の要件に該当する場合
             * 65歳以上で新たに雇用された者(平成29年1月より)
                 * 要件
                     * 現状31日以上の雇用の見込みがある
                     * 1週間の所定労働時間が20時間以上あること
             * 例外
                 * 法人の役員や個人事業主とその家族は原則、雇用保険の被保険者とはならない
    * 運営
        * 公共職業安定所(ハローワーク)
    * 保険料
        * 事業主と労働者の両方が負担
            * 負担割合(保険料率)は業種により異なる
* 給付の種類
    * 4種類
        * 求職者給付
            * 失業者の求職活動中に支給
            * このうち、ベースとなるのが基本手当(いわゆる失業保険)
        * 就職促進給付
            * 基本手当の支給期間中に再就職した場合に支給
        * 教育訓練給付
            * 労働者の能力向上のために厚生労働大臣指定の教育訓練を受けた場合に支給
        * 雇用継続給付
            * 高齢で働いているものや育児、介護をしているものに支給される

## 求職者給付

* 基本手当の受給条件
    * 定年や自己都合による退職の場合
        * 離職日以前の2年間に被保険者期間が通算して12ヶ月以上あること
    * 倒産や解雇の場合
        * 離職日以前の1年間に被保険者期間が6ヶ月以上あること
    * これらのどちらかに該当し、働く意思と能力はあるが職業に就くことができないこと
* 基本手当の給付日数
    * 倒産、解雇による場合
        * (略)
        * 最大給付日数は330日
    * 自己都合退職、定年退職の場合
        * 被保険者期間
            * 1年未満 : 給付無し
            * 1年以上10年未満 : 90日  (<- 最小給付日数)
            * 10年以上20年未満 : 120日
            * 20年以上 : 150日 (<- 自己都合、定年の最大給付日数)
* 受給期間と待機期間
    * 受給期間
        * 離職の翌日から1年間
            * 出産や育児で仕事ができない場合は最長3年延長して4年間まで可能
        * 受給期間を過ぎると所定の給付日数が残っていても基本手当は支給されない
    * 待機期間
        * 倒産、解雇の場合
            * ハローワークで受給の申込みがあった後、7日間の待機期間があり、この間は基本手当は支給されない
        * 自己都合の場合
            * 7日間の待機期間後、さらに3ヶ月の待機期間がある(合計3ヶ月の7日間は基本手当は支給されない)
    * 厚生年金を受給できる65歳未満のものが雇用保険の基本手当を受給する場合
        * 雇用保険の受給が優先され厚生年金は全額支給停止になる
        
## 雇用継続給付

* 種類
    * 高年齢雇用継続給付
    * 育児休業給付
        * 満1歳未満の子供が対象
    * 介護休業給付
