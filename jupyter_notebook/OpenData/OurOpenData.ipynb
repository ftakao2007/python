```python
csv_url = 'csv/H29-05-01.csv'
df = pd.read_csv(csv_url, skiprows=4, encoding="SHIFT-JIS")
df = df.iloc[1:25,[0,2,3]]
df.columns = [u'市町村', u'男', u'女']
df[u'男'] = df[u'男'].str.replace(',','')
df[u'男'] = pd.to_numeric(df[u'男'])
df[u'女'] = df[u'女'].str.replace(',','')
df[u'女'] = pd.to_numeric(df[u'女'])
#df = df.convert_objects(convert_numeric=True)
df = df.set_index(u'市町村')
df
```
