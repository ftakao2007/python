## 利益の種類

5種類ある

### 売上総利益

<img src="img/pl3.jpg" alt="売上総利益" title="売上総利益" width="1000" height="800" align="left" /> 

<br clear="left">

* 本業の利益
    * 売上高 - 売上原価
    * 一番会社の特徴が反映される利益
    * 株の売買等は本業ではないのでここには反映されない

### 営業利益

<img src="img/pl4.jpg" alt="営業利益" title="営業利益" width="1000" height="800" align="left" /> 

<br clear="left">

* 売上総利益から販管費を引いたもの
    * 販管費は売上総利益を獲得するためにどれだけ企業が努力したかを示す
        * 広告料、人件費、店舗の賃料がここに入る
        * 販管費の大小で良い悪いは一概には言えない
        
### 経常利益

<img src="img/pl5.jpg" alt="経常利益" title="経常利益" width="1000" height="800" align="left" /> 

<br clear="left">

* 会社の実力が一番反映される利益
    * 本業で獲得した利益に、本業以外で獲得した収益と費用を加算して算出される
        * 株の売買
        * 利息収支
        
### 税引前当期純利益

<img src="img/pl6.jpg" alt="税引前当期純利益" title="税引前当期純利益" width="1000" height="800" align="left" /> 

<br clear="left">

* １期間に発生したすべての事象を加味して算出された利益
    * 特別損益
        * 毎期発生しない事象から発生した損益も含む
            * 火災損失
            * 事業の売却
            
### 当期純利益

* 税引前当期純利益に税金コストを加味したもの
